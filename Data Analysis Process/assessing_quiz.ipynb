# Assessing and Building Intuition Quiz
Use the space below to explore `census_income_data.csv` to answer the quiz questions below.

```python
# Import pandas
import pandas as pd
import numpy as np

# Load census income data
df = pd.read_csv('census_income_data.csv')
df.head()
```

```python
# Work to answer the quiz question

# Q1. Number of Rows
print("\nNumber of Rows: ",df.shape[0])

# Q2. Number of Columns
print("\nNumber of Columns: ",df.shape[1])

# Q3. Data Type of Several Columns
print("\nData Types: \n",df.dtypes)

# Q4. Columns with Missing Values
print("\nColumns with Missing Values: \n",df.columns[df.isnull().any()])

# Q5. Number of Unique Values of Education
print("\nNumber of Unique Values of Education: ",len(np.unique(df['education'])))
print("\nUnique Values in Education are: \n",np.unique(df['education']))

# Q6. Mean Age
print("\nMean Age: ",df['age'].mean())

# Q7. 75th Percentile of Hours per week
print("\n75th Percentile of Hours per Week: ",df['hours-per-week'].quantile(q=0.75))
```
